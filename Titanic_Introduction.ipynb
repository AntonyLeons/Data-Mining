## Head
Head allows us to take the 'top' of the dataframe object. Has a default value, but can specify an integer.

```python
data.head(15)
```

```python
# Methods and Attributes of a DataFrame are available here:
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
# Documentation is important!

subData = data.head(15)
number_of_headings = 0

for heading in subData.columns: # Similar to collections in C#.
    print(heading)
    number_of_headings += 1

print(f"Number of headings: {number_of_headings}") # f-strings
print("Number of headings(2): " + str(number_of_headings)) # String concatenation

print("\n\n\n") # Printing newlines.

# Iterate over some collection.
for entry in subData.values:
    print(entry)
```

# Complex querying
DataFrames are very versatile, allowing you to select only attributes you want.
The syntax for this is similar to array indexing. However, we pass it a string-array of columns we want to use.
Sample can be used to randomly select data entries from our full data.

```python
data[['PassengerId', 'Survived']].sample(50)
```

### Useful for figuring out number of patterns for useful attributes.
This is how I arrived at 0.42 for min age in the lecture slides, and the maximum age.

```python
data[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].describe()
```
